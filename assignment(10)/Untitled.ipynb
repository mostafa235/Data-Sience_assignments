```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import sqlite3
```

# fill books tabel

```python
soup=BeautifulSoup(requests.get('https://books.toscrape.com/').text,'html.parser')
d={'Zero':0,'One':1,'Two':2,'Three':3,'Four':4,'Five':5}

for li_cat in soup.find('div',attrs={'class':'side_categories'}).find('ul',attrs={"class":"nav nav-list"}).find('ul').find_all('li'):
    soup_cat=BeautifulSoup(requests.get('https://books.toscrape.com/'+li_cat.find('a').get('href')).text,'html.parser')
    for li in soup.find_all('article',attrs={'class':'product_pod'}):
        title=li.find('h3').find('a').get('title')
        rate=d[li.find('p').get('class')[1]]
        price=float(li.find('p',attrs={'class':'price_color'}).get_text()[2:])
        connections = sqlite3.connect('scrap.db')
        connections.execute('insert into books (BookTitle , Price ,Rate ) values (?,?,?)' ,[title,price,rate])
        connections.commit()
```

# fill categories tabel 

```python
soup=BeautifulSoup(requests.get('https://books.toscrape.com/').text,'html.parser')
d={'Zero':0,'One':1,'Two':2,'Three':3,'Four':4,'Five':5}

for li_cat in soup.find('div',attrs={'class':'side_categories'}).find('ul',attrs={"class":"nav nav-list"}).find('ul').find_all('li'):
    soup_cat=BeautifulSoup(requests.get('https://books.toscrape.com/'+li_cat.find('a').get('href')).text,'html.parser')
    category=li_cat.get_text().strip()

    for li in soup.find_all('article',attrs={'class':'product_pod'}):
        title=li.find('h3').find('a').get('title')
        connections = sqlite3.connect('scrap.db')
        connections.execute('insert into Categories ( BookTitle ,Category ) values (?,?)' ,[title,category])
        connections.commit()
```
