```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('loan_data.csv')
df.head(10)
```

```python
sns.pairplot(df)
```

```python
sns.kdeplot(df['LoanAmount'])
```

```python
sns.set(rc={"figure.figsize":[10,10]},font_scale=1.3)

sns.jointplot(df['ApplicantIncome'],df['LoanAmount'],kind="scatter")
```

```python
sns.set(rc={"figure.figsize":[10,10]},font_scale=1.3)

sns.stripplot(df['Gender'],df['LoanAmount'])
```

```python
sns.set(rc={"figure.figsize":[10,10]},font_scale=1.3)
sns.violinplot(df['Education'],df['LoanAmount'])
```

```python
sns.boxplot(df['Married'],df['ApplicantIncome'])
```

```python
sns.barplot(df['Self_Employed'],df['LoanAmount'])
```

```python
sns.countplot(df['LoanAmount'])
```

```python
sns.barplot(x='Self_Employed',y='LoanAmount',data=df,estimator=np.max)
```

```python
sns.barplot(df['Education'],df['LoanAmount'],estimator=np.mean)
```
