# task 4

```python
df=pd.read_csv('visitor-interests.csv')
```

```python
df
```

```python
df=df[:50]
```

```python
df.info()
```

```python
df
```

```python
import user_agents 
def extract(ua):
    ua=user_agents.parse(ua)
    return pd.Series((ua.os.family,ua.browser.family,ua.device.family))
df[['os','browser','device']]=df['UserAgent'].apply(extract)
df.head()
```

```python
from ip2geotools.databases.noncommercial import DbIpCity as ip2geo
def Ip_extract(ip):
    try:   
        ua=ip2geo.get(ip,api_key='free')
        return pd.Series((ua.latitude,ua.longitude))

    except:
        return 'missing'

df[['latitude','longitude']]=df['IP'].apply(Ip_extract)
df.head()
```

# task 5

```python
from geopy.distance import great_circle
```

```python
def extract_Dist_From_Egypt_from_ip(ip):
    try:
        ua = ip2geo.get(ip, api_key='free')
        
        return great_circle((ua.latitude,ua.longitude),(30.06,31.25)).kilometers
    except:
        return 'Missing'

df['Dist_From_Egypt']=df['IP'].apply(extract_Dist_From_Egypt_from_ip)
df.head()
```
