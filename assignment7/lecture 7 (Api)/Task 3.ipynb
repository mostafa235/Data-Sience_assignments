```python
import csv   
```

```python
status=0
```

```python
with open ('user data.csv','w',newline='') as f:
    writer=csv.DictWriter(f,fieldnames=['id','name','email','gender','status'])
    writer.writeheader()
    for user in data['data']:
        id=user['id']
        name=user['name']
        email=user['email']
        gender=user['gender']
        if user['status']=='inactive':
            status=0
        else: status=1
        writer.writerow({'id': id,'name':name,'email':email,'gender':gender,'status':status})
```
