```python
import csv
from  dateutil.parser import parse
```

```python
from datetime import datetime
```

```python
with open('weather.csv','a',newline='') as f:
    writer=csv.DictWriter(f,fieldnames=['date','time','temperature','humidity','visibility','wind_speed','air_pressure'])
    #writer.writeheader()
    for weather in data["consolidated_weather"]:
        date=parse(weather['applicable_date']).date()
        
        n=parse(data["time"])
        x=str(n.hour)
        x1=str(n.minute)
        x2=str(n.second)
        x3=x+":"+x1+":"+x2
        x3
        time=datetime.strptime(x3,"%H:%M:%S").time()
        
        temperature=weather['the_temp']
        humidity=weather['humidity']
        visibility=weather['visibility']
        wind_speed=weather['wind_speed']
        air_pressure=weather['air_pressure']
       
        writer.writerow({'date':date,'time':time,'temperature':temperature,'humidity':humidity,'visibility':visibility,'wind_speed':wind_speed,'air_pressure':air_pressure})
        
```
